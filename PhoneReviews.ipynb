#Siplifying star reviews
Adding a good, bad or medium rating column. 

```python
def convert_rating(star_rating):
  if star_rating == 1:
    return ('1')

  if star_rating == 2:
    return ('2')
  
  if star_rating == 3:
    return ('3')

  if star_rating == 4:
    return ('4')
  
  if star_rating == 5:
    return ('5')

good_or_bad_list = []

for rating in reviews["rating"]:
  good_or_bad_list.append(convert_rating(rating))

reviews["GoodOrBad"] = good_or_bad_list
reviews.head()
```

#Preparing the data
Spacy requires the data to be presented to it in a certain way. The code is formated as it requires.

```python
texts = reviews.body.values

labels = []
for item in reviews.GoodOrBad.values:
  label = {'1 star': item == '1',
           '2 stars': item == '2',
           '3 stars': item == '3',
           '4 stars': item == '4',
           '5 stars': item == '5'}
  labels.append(label)

train_text = texts
train_labels = []
for label in labels:
  categories = {"cats": label}
  train_labels.append(categories)

  
#for text, label in zip(texts[:10], labels[:10]):
  #print(text, label)
```

#Creating the model

```python
nlp = spacy.blank('en')
textcat = nlp.create_pipe("textcat", config={"exectutive_classes": True,
                                             "architecture": "bow"})

nlp.add_pipe(textcat)
textcat.add_label('1 star')
textcat.add_label('2 stars')
textcat.add_label('3 stars')
textcat.add_label('4 stars')
textcat.add_label('5 stars')
```

```python
losses = {}
train_data = list(zip(train_text, train_labels))
batches = minibatch(train_data[:1000], size=10)
for batch in batches:
  texts, labels = zip(*batch)
  nlp.update(texts, labels, sgd=nlp.begin_training(), losses=losses)

```

#Creating the review


```python
test_review = input('Your review: ')
print("The decimal points below show the likelyhoods that the review would be given that number of stars.")
doc = nlp(test_review)
print(doc)
print(doc.cats)
```
