```
import pandas as pd
import matplotlib.pyplot as plt
```

Today we are looking at global temperatures. We will be looking at yearly averages to see if global warming is happening.

```
temperatures = pd.read_csv('GlobalTemperatures.csv')
temperatures.head()
```

```
temperatures.tail()
```

```
reduced_data = temperatures.iloc[:,:2]
reduced_data.head()
```

Now that the data has been loaded I will calculate a few averages before plotting.

```
mean = reduced_data['LandAverageTemperature'].mean()
std = reduced_data['LandAverageTemperature'].std()
print(f'the mean temperature through the years was: {mean}')
print(f'the average deviation was: {mean}')
```

```
times = pd.DatetimeIndex(reduced_data['dt'])
yearly_mean = reduced_data.groupby([times.year]).mean()
yearly_std = reduced_data.groupby([times.year]).std()
```

```
plt.figure(figsize=(20, 10))
plt.plot(yearly_mean['LandAverageTemperature'])
plt.title('Average temperature between 1750 - 2015')
plt.xlabel('year')
plt.ylabel('temperature (Â°C)')
plt.show()
```

```
new_data = temperatures[['dt', 'LandAverageTemperatureUncertainty']]
yearly_uncertainty = new_data.groupby([times.year]).mean()
plt.figure(figsize = (20, 10))
plt.plot(yearly_uncertainty['LandAverageTemperatureUncertainty'])
plt.title('Average Uncertainty')
plt.xlabel('year')
plt.ylabel('temperature(Â°C)')
plt.show()
```

The data shows that temperatures are going up whilst uncertainty is going down.
